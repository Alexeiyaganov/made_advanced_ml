# MCMC sampling based on bigrams

```python
dictionary = {i: ix for ix, i in enumerate(sorted([i for i in set(rus) if re.search('[а-я ]', i)]))}
```

```python
trans_probs = np.ones((len(dictionary), len(dictionary)))

for l, ix in rus_counter2.most_common()[1:]:
    trans_probs[dictionary[l[1]], dictionary[l[0]]] = ix
```

```python
dic = {} # dictionary: index of symbol -> # of times it occurs in the text
for k in range(len(dictionary)):  
    dic[k] = np.sum(trans_probs[k])

for l in range(trans_probs.shape[0]):
    for m in range(trans_probs.shape[1]): 
        trans_probs[l,m] /= (dic[l])
```

```python
def log_like(rus_decode):
    
    log_like = 0
    for i in range(1, len(rus_decode)):
        if (rus_decode[i-1] in dictionary and rus_decode[i] in dictionary):
            index_1 = dictionary[rus_decode[i-1]]
            index_2 = dictionary[rus_decode[i]]
            
            if (trans_probs[index_2,index_1]) > 0:
                log_probs = np.log(trans_probs[index_2,index_1])
                log_like += log_probs
    
    return log_like 
```

```python
log_like(mess_pred)
```

```python
def MC(max_iterations=1000):
    current_permutation = {i: j for i, j in zip(list(dictionary.keys()), np.random.permutation(list(dictionary.keys())))}
    
    for i in range(max_iterations): 
        index_1 = np.random.choice(list(dictionary.keys()))
        index_2 = np.random.choice(list(dictionary.keys()))
        
        while (index_1 == index_2):
            index_2 = np.random.choice(list(dictionary.keys()))
        
        new_permutation = current_permutation.copy()
        
        new_permutation[index_1], new_permutation[index_2] = new_permutation[index_2], new_permutation[index_1]
        
        current_message = ''.join([current_permutation.get(i, i) for i in rus_code])
        prob_message = ''.join([new_permutation.get(i, i) for i in rus_code])
        log_accept = log_like(prob_message) - log_like(current_message)
        
        # draw a sample form a uniform distribution over [0,1] to accept or reject the new permutation  
        if (np.random.random() < np.exp(log_accept)):
            current_permutation = new_permutation 
         
            if (i%1000 == 0):
                print("Number of iterations: " + str(i))
                print("Log-likelihood: " + str(log_like(current_message))) 
                print(current_message)
```
